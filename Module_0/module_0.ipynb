```python
import random 

def guess (lim_low,lim_high):##процедура отгадывания случайного числа на заданном диапазоне 
 b=[0,0]
 x= (random.randint(lim_low,lim_high))
 print(f'компьютер загадал число {x}')
 b[0]=lim_low ##задаем нижнюю границу 
 b[1]=lim_high ##задаем верхнюю границу
 count=0 ## инициализируем счетчик
 while x not in b: ## делим диапазон пополам до тех пора пока загаданное число не попадед на границу диапазона
   c=(b[0]+b[1])//2 
   if x in range (b[0],c): ##определяем где относительно границы находится число и определяем новые границы
      b[1]=c 
   else :
      b[0]=c
   count+=1     
 if x==b[0]: ## проверяем верхней или нижней границы диапазона равно загаданное число
    return b[0],count+1 
 else:
    return b[1],count+1

##вводим диапазон загадывания числа
low=int(input("введите нижний лимит"))
high=int(input("введите верхний лимит"))
repeat = int(input("число повторов тестирования"))
## тестируем функцию на заданном числе повторений
count_total=[]    
for i in range (repeat):
        s,m=guess(low,high)
        count_total.append(m)
        print (f'алгоритм отгадал число {s} за {m} попыток')
print (f'среднее число попыток {sum(count_total,0)/i}')
```
