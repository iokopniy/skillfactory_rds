# Предобработка

```python
#Альертнативная реализация ф-ции explode
def new_df (df,columns_name):
  newdf=pd.DataFrame(columns=columns_name)
  for i in range (0, len (df)):
    new_series=df.iloc[i][columns_name[1]].split('|')
    for j in range (0, len (new_series)):
      newdf=newdf.append({columns_name[0]:data.iloc[i].imdb_id,columns_name[1]:new_series[j]},ignore_index=True)
  return newdf
#превращение строки с разделителем в массив
def create_list (a):
 return a.split('|')
#получение месяца из строки
def get_month (x):
 return datetime.datetime.strptime(x, '%m/%d/%Y').month 
#получение попарного сочетяния актеров
def duets_creation (cast):
 duets=list(itertools.combinations(cast,2))
 return (duets)


answers = {} # создадим словарь для ответов
#создание дополнительных полей для обработки
data['profit']=data.revenue-data.budget
data['director_list']=data['director'].apply(create_list)
data['genres_list']=data['genres'].apply(create_list)
data['cast_list']=data['cast'].apply(create_list)
data['companies_list']=data['production_companies'].apply(create_list)
data['release_month']=data['release_date'].apply(get_month)
data['name_len']=data['original_title'].apply(len)
data['overview_len']=data['overview'].apply(len)
data['duets_list']=data.cast_list.apply(duets_creation) #пары актеров , снимающихся вместе
display(data)

# тут другие ваши предобработки колонок например:

#the time given in the dataset is in string format.
#So we need to change this in datetime format
# ...
```
